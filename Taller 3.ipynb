## Entonces,  ¿Existen buenas practicas para visualizaciones?

Antes de entrar en detalle con buenas prácticas consideremos lo siguiente:

1) Nuestros ojos no siguen un orden especifico a no ser que sea algo destacable.

2) Es dificil distinguir entre ángulos, áreas y figuras en 3D.

3) Tambien es dificil digerir mucha información en un gráfico (pero tampoco es bueno poner muy poca).

4) Siempre tener en cuenta, que la gente busca "significado" en los datos.

5) Tener en cuenta las convenciones culturales de las personas a la que vaya dirigida el gráfico.

6) Si se tiene muchos datos en un gráfico y es necesaria su visualización, es mejor separar estos en categorias y hacer el mismo gŕafico dinamico, es decir, considerar filtros de sus categorias ó simplemente un espacio muestral que sea representativo.

7) Si se debe presentar un dato en partícular, simplemente se debe mostrar, como por ejemplo, horas trabajadas: 13. 
                                                        
8) Nunca subestimar el potencial que puede tener una tabla.

9) jamás olvidar, que relación no implica causalidad.

Lo que se resume en

a) Preocuparse del contenido del mensaje de los gŕaficos (hacerlo llamativo) y las técnicas que emplearemos para visualizar.

b) Ser simple, pero no tanto.

c) Los datos pueden mentir, por lo que siempre es bueno dejar el claro el contexto de la información y lo que se asume de ella.

Para finalizar, discutamos si el siguiente gráfico representa bién la información que se quiere entregar sore el consumo del tabaco.
